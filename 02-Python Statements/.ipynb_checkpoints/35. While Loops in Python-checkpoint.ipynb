```python
# Syntax of a while loop
# while some_boolean_condition:
    # do something
# else:
    # do something different
```

```python
x = 0

while x < 5:
    print(f'The current value of x is: {x}')
    x = x + 1
else:
    print('X IS NOT LESS THAN 5')
```

```python
x = 50

while x < 5:
    print(f'The current value of x is: {x}')
    x += 1
else:
    print('X IS NOT LESS THAN 5')
```

```python
# 3 keywords - break, continue, pass
# Pass
x = [1,2,3]

for item in x:
    pass

print ('end of my script')
```

```python
# Continue
mystring = 'Sammy'

for letter in mystring:
    if letter == 'a':
        continue
    print(letter)
```

```python
# Break

for letter in mystring:
    if letter == 'a':
        break
    print(letter)
```

```python
x = 0

while x < 5:
    if x == 2:
        break
    print(x)
    x += 1
```
